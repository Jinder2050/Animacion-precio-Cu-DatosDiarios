##**Librerias Necesarias**

```python
from google.colab import files
import pandas as pd
import io
from ipywidgets import interact
data = files.upload()
```

##**Mostrar Dataframe**

```python
df = pd.read_csv('Datos históricos Futuros cobre.csv', decimal=',', sep=';')
df
```

***Mostrando Columnas***

```python
df.columns
```

##**Eliminar columnas innecesarias**

```python
dff = df.drop(['Apertura', 'Máximo', 'Mínimo', 'Vol.', '% var.'], axis = 1)
dff
```

```python
dff.info()
```

```python
dff.isnull().sum()
```

##**Cambiar tipo de dato**

```python
#dff['Fecha'] = pd.to_datetime(dff['Fecha'],format='%d.%m.%Y',  infer_datetime_format=True, errors='coerce', utc=True).dt.strftime('%Y-%m-%d')
dff['Fecha'] = pd.to_datetime(dff['Fecha'], format='%d.%m.%Y')
dff
```

##**Crear una columna año**

```python
dff['Año'] = pd.DatetimeIndex(dff['Fecha']).year
dff
```

```python
dff.info()
```

```python
dff = dff[['Año','Fecha','Último']]
dff
```

##**Groupby**

***Agrupando por Año/Mes y obteniendo el promedio mensual***

```python
#df2 = dff.groupby([dff['Fecha'].dt.year, dff['Fecha'].dt.month])
dff = dff.groupby([dff['Fecha'].dt.year, dff['Fecha'].dt.month]).mean()
dff.head(13)
```

```python
dff.info()
```

***Resumen por año***

```python
dff.index = dff.index.rename(['Año', 'Mes'])
@interact
def sumary(Periodo = list(dff['Año'].astype(int).value_counts().sort_index().index)):
    x = dff[dff['Año'] == Periodo]
    x.drop('Año', axis=1, inplace=True)
    print(x[:])
```

***Reset_index***

```python
dffinal = dff.reset_index(drop=True)
dffinal
```

***Convertir columna 'Año' a datetime***

```python
dffinal['Año'] = pd.to_datetime(dffinal['Año'], format='%Y',  infer_datetime_format=True, errors='coerce').dt.to_period('Y')
dffinal
```

```python
dffinal.info()
```

##**Transponer dataframe**

```python
dffinal = dffinal.transpose()
dffinal
```

##**Exportar datos formato csv**

```python
dffinal.to_csv('CuOutput.csv', encoding='utf-8-sig', header=False)
files.download('CuOutput.csv')
```
